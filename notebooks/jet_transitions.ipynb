## Load and Plot Reference Trajectory

```python
truth_path  = utils.which_os+'Beta_Plane_Jets/data/test_data/csv_files/'
test_data_1 = np.genfromtxt(truth_path+f'0_{0}_{100}_umean.csv', delimiter=',')
for t in range(200, 5100, 100):
    test_data_1 = np.concatenate((test_data_1, np.genfromtxt(truth_path+f'0_{t-100}_{t}_umean.csv', delimiter=',')), axis=1)

start = 0
end   = 5000
plt.figure(figsize=(22, 1.5))
plt.imshow(test_data_1[::-1, start:end], aspect='auto', origin='lower')
for i in range(start, end, 100):
    plt.axvline(x=i, linestyle='--', color='w')
plt.show()
```

# Plot Coalescence Probabaility 250 timesteps from event

```python
event = 1328

f = plt.figure(figsize=(10, 7))
s = 1000
t = 78
arr_ML_1200_2405 = np.load(f'/home/is500/Documents/Beta_Plane_Jets/Beta_Plane_ML/notebooks/ML_{s+t}_100000.npy')

q_5  = np.percentile(arr_ML_1200_2405, 5 , axis=0)
q_95 = np.percentile(arr_ML_1200_2405, 95, axis=0)
q_25 = np.percentile(arr_ML_1200_2405, 25  , axis=0)
q_75 = np.percentile(arr_ML_1200_2405, 75  , axis=0)

plt.axvline(x=q_5 , color='orange', linestyle='--')
plt.axvline(x=q_95, color='orange', linestyle='--')
plt.axvline(x=q_25, color='blue', linestyle='--')
plt.axvline(x=q_75, color='blue', linestyle='--')
plt.axvline(x=event-(s+t), color='grey', linestyle='--')

plt.hist(arr_ML_1200_2405, bins=(520-10)//2, density=True, range=(10,520), histtype='step', linestyle=('solid'), label=f'ML, Ensemble size: {1000000}', color='red')
locs, labels = plt.xticks() 
plt.title(f'{event-(s+t)} time steps from reference merging event. 50% of probabilty mass lies within {q_75-q_25} time steps. 90% of probabilty mass lies within {q_95-q_5} time steps.')
plt.legend(loc='best')
plt.xlabel('time to merging event')
plt.ylim(0, 0.012)
plt.show()
```
